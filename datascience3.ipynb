```python


import pandas as pd
import matplotlib.pyplot as plt


file_path = "/content/Temp.csv"
data = pd.read_csv(file_path, encoding='ISO-8859-1')


data['Min Temp'] = data['Min Temp'].str.replace('C', '').astype(float)
data['Avg Temp'] = data['Avg Temp'].str.replace('C', '').astype(float)
data['Max Temp'] = data['Max Temp'].str.replace('C', '').astype(float)
data['Precipitation'] = data['Precipitation'].str.replace(' mm', '').astype(float)

data['Date'] = pd.to_datetime(data['Date'], format='%b %d, %Y')


plt.figure(figsize=(12, 6))
plt.plot(data['Date'], data['Min Temp'], label='Min Temp (C)', marker='o')
plt.plot(data['Date'], data['Avg Temp'], label='Avg Temp (C)', marker='o')
plt.plot(data['Date'], data['Max Temp'], label='Max Temp (C)', marker='o')

plt.xlabel('Date')
plt.ylabel('Temperature (C)')
plt.title('Temperature Trends in Chennai (Past 15 Days)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()


trend_summary = {
    'Min Temp Trend': 'Increasing',
    'Avg Temp Trend': 'Increasing',
    'Max Temp Trend': 'Increasing',
    'Precipitation': 'Constant at 0 mm'
}


print("Temperature and Precipitation Analysis (Past 15 Days)")
for key, value in trend_summary.items():
    print(f"{key}: {value}")


print("\nInsights gained:")
print("- Gradual warming trend, likely seasonal.")
print("- No significant spikes or precipitation events.")
print("- Temps may continue rising by ~0.2-0.5C daily.")
print("- Short data period, no humidity or wind data for deeper analysis.")

```
